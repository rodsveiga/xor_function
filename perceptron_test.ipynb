# Perceptron

The class `Perceptron` is designed to stop the training once the weights are such that the perceptron is able to linearly separe. It does not seek an optimal hyperplane.

```python
import numpy as np
from perceptron import Perceptron
```

### AND

```python
X = np.array([[0, 0], [0, 1],
             [1, 0], [1, 1]])

y = np.array([0, 0,
              0, 1])
```

```python
percep = Perceptron(X, y)
percep.train(plot_lines= True)
```

```python
percep.epoch_stop()
```

```python
percep.plot()
```

### OR

```python
X = np.array([[0, 0], [0, 1],
             [1, 0], [1, 1]])

y = np.array([0, 1,
              1, 1])
```
